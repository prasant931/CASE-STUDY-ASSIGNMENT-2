```python
import yfinance as yf
import sqlite3

# Define the ticker symbol for Bitcoin
btc_ticker = "BTC-USD"

# Create a Yahoo Finance object
btc_data = yf.Ticker(btc_ticker)

# Retrieve historical data
btc_history = btc_data.history(period="1y")  # You can change the period as needed

# Create a SQLite database
conn = sqlite3.connect("btc_data.db")
cursor = conn.cursor()

# Create a table to store BTC data
cursor.execute("""
    CREATE TABLE IF NOT EXISTS bitcoin_data (
        Date TEXT,
        Open REAL,
        High REAL,
        Low REAL,
        Close REAL,
        Volume INT,
        Dividends REAL,
        Stock_Splits REAL
    )
""")

# Insert BTC data into the database
for index, row in btc_history.iterrows():
    cursor.execute("""
        INSERT INTO bitcoin_data (Date, Open, High, Low, Close, Volume, Dividends, Stock_Splits)
        VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    """, (str(index), row['Open'], row['High'], row['Low'], row['Close'], row['Volume'], row['Dividends'], row['Stock Splits']))

# Commit the changes and close the connection
conn.commit()
conn.close()
```
